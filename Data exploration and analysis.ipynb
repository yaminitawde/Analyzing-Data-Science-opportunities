#### Step 3: Data pre-processing

```python
print(ds_openings.isnull().sum())
```

- On checking for null values we see all columns have null values.

```python
nan_values = ds_openings[ds_openings['position'].isna()]
print(nan_values)
```

```python
print(ds_openings.isnull().sum())
```

#### Step 1:Identifying the key column for merging the data :

- On studying both the datasets we see that **'job state'** column can be used to map a working professional to the job openings available in the same state so we will it as a key for joining both the datasets.

```python
city =[]
state =[]
for x in ds_openings['location'] :
    temp = x.split(",")
    city.append(temp[0])
    state.append(temp[1].strip().split(" ")[0])
ds_openings['city'] = city
ds_openings['state'] = state
ds_openings.head()
```

- The second dataset do not have state column for joining it with first dataset. So we split the location column to retrieve city and state information.

```python
# Renaming columns for consistency
ds_openings.rename(columns={"state": "job_state","position":"open_position","company":"hiring_company","description":"requirement_desc"
                           ,"reviews":"hiring_cmp_rvw","location":"hiring_location","city":"hire_job_city"}, inplace= True)
```

#### Step 2: Combining the two datasets 

We are using left join as we want to maintain the working professionals data i.e left table irrespective of his/her location not having job openings i.e right table. This will help us to gain an comprehensive view about availability as well as non-availability of a new job for the working professional in the same state.

```python
result = pd.merge(glassdoor_listing,ds_openings, how="left", on=["job_state"])
```

#### Step 3: Checking the resulting data

```python
result.head()
```

#### Step 4: Aggregating based on simplified job title and location

We aggregate the joined data by simplied job title and location to create a new dataframe containing the job count available for the specific job title at a particular location. We will use this aggregated data to further draw insights and for visualization.

```python
job_count = result.groupby(['job_simp','Location']).count()['job_state'].reset_index(name="job_count")
print(job_count)
```

> Salary ranges in different states

```python
#Find the mean average salary, minimum salary and maximum salary in different states
avg_sal = result.groupby(['job_state']).mean()['avg_salary']
min_sal = result.groupby(['job_state']).mean()['min_salary']
max_sal = result.groupby(['job_state']).mean()['max_salary']

state = max_sal.index
```

```python
fig=plt.figure(figsize = (10,6))
ax=fig.add_axes([0,0,1,1])
ax.scatter(state, max_sal)
ax.scatter(state, avg_sal)
ax.scatter(state, min_sal)

ax.set_xlabel('States')
ax.set_ylabel('Salary range')
ax.set_title('Salary Range in different states')
plt.legend([ 'Maximum Salary','Average salary', 'Minimum Salary'], fontsize = 'x-large', bbox_to_anchor=(0.6, 0.9), fancybox=True, shadow=True)
plt.show()
```

```python
bars = np.add(min_sal, avg_sal).tolist()
plt.figure(figsize = (18,10))
plt.bar(state, min_sal)
plt.bar(state, avg_sal, bottom=min_sal)
plt.bar(state, max_sal, bottom=bars)

plt.xlabel('States')
plt.ylabel('Salary range')
plt.title('Stack of Salary Range in different states')
plt.legend([ 'Minimum Salary','Average salary', 'Maximum Salary'], fontsize = 'x-large', loc="upper right", fancybox=True, shadow=True)
plt.show()

# Ref: https://www.python-graph-gallery.com/12-stacked-barplot-with-matplotlib
```

- We observe that the state of Delaware(DE) has one of the highest range of salaries paid. The average salary paid by companies in Delaware is double than most of what companies are paying in other states. 

- From the previous bar plots we observe that the number of job listings in Delaware is very low as compare to most of the listings in other states, still we observe that the salary range is very high. Why is it so? Let's analyze more!!!

> Salary ranges across different sectors

```python
# Grouping Job State with Sector and finding the mean salary
job_sect = result.groupby(['job_state','Sector']).mean()['avg_salary'].reset_index(name="avg_sal")

#Finding avg salary for different states
#We are selecting a mix of different states based on salary ranges as observed in the stacked bar plot
state_ca = job_sect[(job_sect.job_state == "CA")]
state_de = job_sect[(job_sect.job_state == "DE")]  #State with the highest salary range
state_nc = job_sect[(job_sect.job_state == "NC")]
state_ma = job_sect[(job_sect.job_state == "MA")]

#Finding avg salary for different sectors
#We will select the states which have lot Sectors 
sec_insurance = job_sect[(job_sect.Sector == "Insurance")]
sec_bio = job_sect[(job_sect.Sector == "Biotech & Pharmaceuticals")]
sec_health = job_sect[(job_sect.Sector == "Health Care")]
sec_it = job_sect[(job_sect.Sector == "Information Technology")]
```

```python
fig , ax = plt.subplots(2,2, figsize=(14,10))
fig.tight_layout(pad=8.0) #padding space between subplots

ax[0][0].plot(sec_insurance['job_state'], sec_insurance['avg_sal'], marker='o')
# ax[0][0].grid(True)
ax[0][0].tick_params(labelrotation=45)
ax[0][0].set_title('Avg Salary in Insurance sector', fontsize=14)
ax[0][0].set_xlabel('States', fontsize=14)
ax[0][0].set_ylabel('Salary range', fontsize=14)


ax[0][1].plot(sec_bio['job_state'], sec_bio['avg_sal'], marker='o')
# ax[0][1].grid(True)
ax[0][1].tick_params(labelrotation=45)
ax[0][1].set_title('Avg Salary in Biotech & Pharmaceuticals sector', fontsize=14)
ax[0][1].set_xlabel('States', fontsize=14)
ax[0][1].set_ylabel('Salary range', fontsize=14)


ax[1][0].plot(sec_health['job_state'], sec_health['avg_sal'], marker='o')
#ax[1][0].grid(True)
#ax[1][0].tick_params(labelrotation=45)
ax[1][0].set_title('Avg Salary in Healthcare sector', fontsize=14)
ax[1][0].set_xlabel('States', fontsize=14)
ax[1][0].set_ylabel('Salary range', fontsize=14)


ax[1][1].plot(sec_it['job_state'], sec_it['avg_sal'], marker='o')
#ax[1][1].grid(True)
#ax[1][1].tick_params(labelrotation=45)
ax[1][1].set_title('Avg Salary in Information Technology sector', fontsize=14)
ax[1][1].set_xlabel('States', fontsize=14)
ax[1][1].set_ylabel('Salary range', fontsize=14);

# Reference: https://www.tutorialspoint.com/matplotlib/matplotlib_subplots_function.htm
```
